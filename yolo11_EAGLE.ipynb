```python
from ultralytics import YOLO
from sklearn.metrics import precision_recall_fscore_support
import numpy as np

# Caricamento modello YOLO pre-addestrato
model = YOLO("yolov5su.pt")

# File di configurazione del dataset
dataset_yaml = 'EAGLE2/dataset.yaml' 

# Funzione per calcolare IoU, precision, recall, F1-score e mAP
def evaluate_metrics(results):
    # Estrai i dati di valutazione
    metrics = results.metrics
    precision = metrics['precision']  # Precisione
    recall = metrics['recall']        # Recall
    f1 = metrics['f1']                # F1-Score
    iou = metrics['iou']              # Intersection over Union (IoU)
    map_value = metrics['mAP_0.5']    # Mean Average Precision (mAP) con threshold 0.5

    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")
    print(f"IoU: {iou:.4f}")
    print(f"mAP (IoU=0.5): {map_value:.4f}")

# Addestramento e valutazione
def train_and_evaluate():
    # Esegui l'addestramento e la valutazione sul validation set
    results = model.train(data=dataset_yaml, epochs=100, imgsz=640)
    print(results)
# Esegui il training e la valutazione
train_and_evaluate()
```

```python
from ultralytics import YOLO
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Carica il modello
from ultralytics import YOLO
from pathlib import Path
import os

# Carica il modello addestrato
model_ = YOLO("runs/detect/train23/weights/best.pt")
#model_ = YOLO("yolo11n.pt")

nome_cartella = "EAGLE2"
dataset_yaml_ = f'{nome_cartella}/dataset.yaml'
print("YAML path:", dataset_yaml_)
assert os.path.exists(dataset_yaml_), "❌ File dataset.yaml non trovato!"


def eval_metrics(results):
    print("==> Risultati della valutazione:")
    print(f"Precisione media (P): {results.box.mp:.4f}")
    print(f"Recall media (R): {results.box.mr:.4f}")
    print(f"mAP@0.5: {results.box.map50:.4f}")
    print(f"mAP@0.5:0.95: {results.box.map:.4f}")

def extract_counts_from_predictions(predictions, labels_dir):
    pred_counts = []
    gt_counts = []

    for pred in predictions:
        pred_counts.append(len(pred.boxes))

        image_stem = Path(pred.path).stem
        label_path = Path(labels_dir) / f"{image_stem}.txt"

        if label_path.exists():
            with open(label_path, 'r') as f:
                lines = f.readlines()
            gt_counts.append(len(lines))
        else:
            gt_counts.append(0)

    return gt_counts, pred_counts



import matplotlib.pyplot as plt
import numpy as np

def plot_counts(gt_counts, pred_counts, title):
    gt_counts = np.array(gt_counts)
    pred_counts = np.array(pred_counts)

    plt.figure(figsize=(8, 6))
    plt.scatter(gt_counts, pred_counts, alpha=0.6, label="Predizioni")
    
    min_val = min(gt_counts.min(), pred_counts.min())
    max_val = max(gt_counts.max(), pred_counts.max())
    plt.plot([min_val, max_val], [min_val, max_val], 'r--', label="Linea ideale")

    plt.xlabel("True Count")
    plt.ylabel("Predicted Count")
    plt.title(f"{title} - Pred vs True")
    plt.grid(True)
    plt.legend()

    # Salva il plot come immagine
    filename = title.lower().replace(" ", "_") + "_pred_vs_true.png"
    plt.savefig(filename, bbox_inches='tight')
    plt.close()



def compute_and_print_stats(gt_counts, pred_counts, set_name):
    mae = mean_absolute_error(gt_counts, pred_counts)
    mse = mean_squared_error(gt_counts, pred_counts)
    r2 = r2_score(gt_counts, pred_counts)

    print(f"\n==> Statistiche di conteggio per {set_name}:")
    print(f"MAE: {mae:.2f}")
    print(f"MSE: {mse:.2f}")
    print(f"R²:  {r2:.2f}")

    plot_counts(gt_counts, pred_counts, f"{set_name} - Ground Truth vs Predicted Counts")

def valuta():
    print("\nValutazione sul validation set:")
    val_results = model_.val(data=dataset_yaml_, save_json=False)
    #eval_metrics(val_results)

    val_preds = model_.predict(source=f"{nome_cartella}/valid/images", save=False)
    #val_gt_counts, val_pred_counts = extract_counts_from_predictions(val_preds, val_results)
    val_gt_counts, val_pred_counts = extract_counts_from_predictions(val_preds, f"{nome_cartella}/valid/labels")

    compute_and_print_stats(val_gt_counts, val_pred_counts, "Validation Set")

    print("\nValutazione sul test set:")
    test_results = model_.val(data=dataset_yaml_, split="test", save_json=False)
    #eval_metrics(test_results)

    test_preds = model_.predict(source=f"{nome_cartella}/test/images", save=False)
    #test_gt_counts, test_pred_counts = extract_counts_from_predictions(test_preds, test_results)
    test_gt_counts, test_pred_counts= extract_counts_from_predictions(test_preds, f"{nome_cartella}/test/labels")
    compute_and_print_stats(test_gt_counts, test_pred_counts, "Test Set")

# Avvia la valutazione
valuta()
```
